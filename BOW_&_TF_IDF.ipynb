```python
import re
import pandas as pd
import numpy as np
```

```python
messages = pd.read_csv('/content/SMSspam.csv', encoding = 'latin-1')
```

```python
messages = messages[['v1','v2']]
```

```python
messages.columns = [['label','message']]
```

```python
messages.head()
```

```python
messages.columns = [col[0] for col in messages.columns]
print(messages.columns)
```

```python
from nltk.stem import SnowballStemmer
```

```python
ss = SnowballStemmer('english')
```

```python
import nltk
import re
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
corpus = []
for i in range(len(messages)):
  review = re.sub('[^A-Za-z]',' ',messages['message'].iloc[i])
  review = review.lower()
  review = review.split()

  review = [ss.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

# BAG OF WORDS

```python
y = pd.get_dummies(messages['label'])
y = y.iloc[:,0].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, ngram_range=(1,2))
```

```python
type(X_test)
```

```python
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
np.printoptions(edgeitems = 30, linewidth = 10000,
                formatter = dict(float = lambda x: ".3g" % x))
X_train
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(f" Accuracy = {accuracy_score(y_test,y_pred)}\n")
print(f" Classification Report \n {classification_report(y_test,y_pred)}")
```
